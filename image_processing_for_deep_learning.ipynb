# **Image resizng using Pillow library

```python
from PIL import Image

img =Image.open('/content/cat.jpg')
img_resized = img.resize((200,200))
```

```python
print(img_resized.size)
```

```python
img_resized.save('cat_resized.jpg')
```

```python
img_res = mpimg.imread('/content/cat_resized.jpg')
plt.imshow(img_res)
plt.show()
```

```python
# Converting the image from RGB to GrayScale
import cv2
```

```python
from cv2.gapi import RGB2Gray
img_to_convert= cv2.imread("/content/cat_resized.jpg")
gray_img =cv2.cvtColor(img_to_convert, cv2.COLOR_RGB2GRAY)

```

```python
# Now to see the gray image we have a function cv2.imshow(), but not directly work in colab,  so we haveto import
from google.colab.patches import cv2_imshow
cv2_imshow(gray_img)
```

```python
# Saving the Image
cv2.imwrite('gray_cat.jpg', gray_img)
```
